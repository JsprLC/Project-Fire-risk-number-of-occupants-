## ALKIS® Actual Use (TN)

```python
import geopandas as gpd

# Read the Shapefile (.shp and associated files)
shapefile_path = '/Users/chupei/Desktop/Geo data/tn_09162/Nutzung.shp'
gdf_nutzung = gpd.read_file(shapefile_path)

# Display basic information
print("=== Shapefile Information ===")
print(f"Number of features: {len(gdf_nutzung)}")
print(f"CRS: {gdf_nutzung.crs}")
print("\nColumns:")
for i, col in enumerate(gdf_nutzung.columns, 1):
    print(f"{i:2d}. {col}")

# Display first few rows
print("\n=== First few records ===")
print(gdf_nutzung.head())
```

```python
# Explore all components
print("=== Complete Shapefile Information ===")
print(f"1. Geometry type: {gdf_nutzung.geom_type.unique()}")
print(f"2. Coordinate system: {gdf_nutzung.crs}")
print(f"3. Number of features: {len(gdf_nutzung)}")

# Attribute information (from .dbf)
print("\n=== Attribute Information ===")
print(gdf_nutzung.info())

# Spatial information (from .shp)
print("\n=== Spatial Information ===")
print(f"Bounding box: {gdf_nutzung.total_bounds}")
print(f"Area covered: {gdf_nutzung.unary_union.area:.2f} square units")

# Data preview with both geometry and attributes
print("\n=== Data Preview (Geometry + Attributes) ===")
print(gdf_nutzung.head())
```
